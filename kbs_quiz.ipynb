This notebook implements a rule-based Knowledge-Based System using forward chaining to determine campus access decisions based on facts and logical rules.

## 1. KBS Engine Implementation

```python
class KnowledgeBase:
    """
    A simple Knowledge-Based System that stores facts and rules,
    and performs forward chaining inference.
    """
    
    def __init__(self):
        """Initialize empty knowledge base with facts and rules."""
        self.facts = set()
        self.rules = []
    
    def add_fact(self, fact):
        """Add a fact to the knowledge base."""
        self.facts.add(fact)
    
    def add_rule(self, rule):
        """Add a rule to the knowledge base.
        
        Args:
            rule: Dictionary with 'condition' (function) and 'conclusion' (string)
        """
        self.rules.append(rule)
    
    def infer(self):
        """Apply all rules and infer new facts using forward chaining.
        
        Returns:
            Set of newly inferred facts
        """
        newly_inferred = set()
        changed = True
        
        # Keep applying rules until no new facts are inferred
        while changed:
            changed = False
            for rule in self.rules:
                # Check if rule condition is satisfied
                if rule["condition"](self.facts):
                    conclusion = rule["conclusion"]
                    # Add conclusion if it's new
                    if conclusion not in self.facts:
                        self.facts.add(conclusion)
                        newly_inferred.add(conclusion)
                        changed = True
        
        return newly_inferred
```

## 2. Initialize Knowledge Base and Add Base Facts

```python
# Create knowledge base instance
kb = KnowledgeBase()

# Add initial facts about a student
kb.add_fact("has_student_id")
kb.add_fact("is_registered_student")
kb.add_fact("has_paid_fees")
kb.add_fact("has_health_clearance")

# Display initial facts
print("Initial facts in knowledge base:")
for fact in sorted(kb.facts):
    print(f"- {fact}")
```

## 3. Add Rules to Knowledge Base

```python
# Rule 1: Single-condition rule
# IF has_student_id THEN identity_confirmed
kb.add_rule({
    "condition": lambda facts: "has_student_id" in facts,
    "conclusion": "identity_confirmed"
})

# Rule 2: Single-condition rule
# IF has_health_clearance THEN health_verified
kb.add_rule({
    "condition": lambda facts: "has_health_clearance" in facts,
    "conclusion": "health_verified"
})

# Rule 3: Multi-condition rule
# IF identity_confirmed AND is_registered_student THEN is_verified_student
kb.add_rule({
    "condition": lambda facts: "identity_confirmed" in facts and "is_registered_student" in facts,
    "conclusion": "is_verified_student"
})

# Rule 4: Multi-condition rule
# IF is_verified_student AND has_paid_fees THEN financial_clearance_granted
kb.add_rule({
    "condition": lambda facts: "is_verified_student" in facts and "has_paid_fees" in facts,
    "conclusion": "financial_clearance_granted"
})

# Rule 5: Chained rule (depends on conclusions from previous rules)
# IF financial_clearance_granted AND health_verified THEN access_requirements_met
kb.add_rule({
    "condition": lambda facts: "financial_clearance_granted" in facts and "health_verified" in facts,
    "conclusion": "access_requirements_met"
})

# Rule 6: Final decision rule (extra rule for originality)
# IF access_requirements_met THEN campus_entry_granted
kb.add_rule({
    "condition": lambda facts: "access_requirements_met" in facts,
    "conclusion": "campus_entry_granted"
})

print(f"Successfully added {len(kb.rules)} rules to the knowledge base.")
```

## 4. Run Inference and Display Results

```python
# Perform inference
new_facts = kb.infer()

# Display results
print("=== INFERENCE RESULTS ===")
print("\nNewly inferred facts:")
for fact in sorted(new_facts):
    print(f"- {fact}")

print("\nFinal facts after inference:")
for fact in sorted(kb.facts):
    print(f"- {fact}")

# Make final decision
print("\n" + "="*40)
if "campus_entry_granted" in kb.facts:
    print("FINAL DECISION: ENTRY GRANTED")
else:
    print("FINAL DECISION: ENTRY DENIED")
print("="*40)
```
